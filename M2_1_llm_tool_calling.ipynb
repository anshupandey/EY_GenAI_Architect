# Tool Calling with Azure OpenAI

```python
from openai import OpenAI
from dotenv import load_dotenv


load_dotenv(override=True)

ENDPOINT   = os.getenv("AZURE_OPENAI_ENDPOINT", "").rstrip("/")
API_KEY    = os.getenv("AZURE_OPENAI_API_KEY")
DEPLOYMENT = os.getenv("AZURE_OPENAI_DEPLOYMENT", "gpt-4o-mini")


BASE_URL   = f"{ENDPOINT}/openai/v1/"

client = OpenAI(api_key=API_KEY, base_url=BASE_URL)
```

### Tool Integration with LLM
- metadata
    - name of tool/function
    - description of the tool/function
    - arguments details - arguments and its description for the tool/function

```python
tools = [{
    "type":"function",
    "function":{
        "name":"get_current_weather",
        "description":"this function can be used to get the current weather information for any given city.",
        "parameters":{
            "type":"object",
            "properties":{"city":{"type":"string","description":"name of city e.g. mumbai, new york"}},
        },
        "required":['city'],
    }}
]
```

```python
tool_map = {"get_current_weather": get_current_weather}

def generate_response(prompt):

    input_messages = [
        {"role": "system", "content": "you are a helpful assistant"},
        {"role": "user", "content": prompt}
    ]

    first_response = client.responses.create(
        model=DEPLOYMENT,
        input=input_messages,
        temperature=0.2,
        tools=tools,
        tool_choice="auto"
    )

    # Separate function calls from text output
    function_calls = [item for item in first_response.output if item.type == "function_call"]

    if function_calls:
        print("LLM decided to trigger tool call", function_calls)

        # Build follow-up input: original messages + all function_call items + their outputs
        follow_up_input = list(input_messages)

        tool_outputs = []
        for tool in function_calls:
            tool_name = tool.name
            tool_args = json.loads(tool.arguments)
            fun_to_ex = tool_map[tool_name]
            tool_output = fun_to_ex(**tool_args)

            # Append the model's function_call item as-is
            follow_up_input.append({
                "type": "function_call",
                "call_id": tool.call_id,
                "name": tool_name,
                "arguments": tool.arguments
            })

            # Collect the result
            tool_outputs.append({
                "type": "function_call_output",
                "call_id": tool.call_id,        # must match call_id above
                "output": json.dumps(tool_output)
            })

        follow_up_input.extend(tool_outputs)

        second_response = client.responses.create(
            model="gpt-4o",
            input=follow_up_input,
            temperature=0.1,
            tools=tools
        )

        return second_response.output_text

    else:
        return first_response.output_text
```

```python
import wikipedia

def get_wikipedia_summary(query:str):
    response = wikipedia.summary(query)
    return response
```

```python
tools = [{
    "type":"function",
    "function":{
        "name":"get_current_weather",
        "description":"this function can be used to get the current weather information for any given city.",
        "parameters":{
            "type":"object",
            "properties":{"city":{"type":"string","description":"name of city e.g. mumbai, new york"}},
        },
        "required":['city'],
    }},
    {
    "type":"function",
    "function":{
        "name":"get_wikipedia_summary",
        "description":"this function can be used to get information about any place, location, person, historical information.",
        "parameters":{
            "type":"object",
            "properties":{"query":{"type":"string","description":"name of any person, event, location, place"}},
        },
        "required":['query'],
    }}
]
```

```python
tool_map = {
    "get_current_weather": get_current_weather,
    "get_wikipedia_summary": get_wikipedia_summary
}

def generate_response(prompt):

    input_messages = [
        {"role": "system", "content": "you are a helpful assistant"},
        {"role": "user", "content": prompt}
    ]

    while True:
        response = client.responses.create(
            model=DEPLOYMENT,
            input=input_messages,
            temperature=0.2,
            tools=tools,
            tool_choice="auto"
        )

        function_calls = [item for item in response.output if item.type == "function_call"]

        if function_calls:
            print("LLM decided to trigger tool call", function_calls)

            # Add all function_call items the model returned into the input
            for tool in function_calls:
                input_messages.append({
                    "type": "function_call",
                    "call_id": tool.call_id,
                    "name": tool.name,
                    "arguments": tool.arguments
                })

            # Execute each tool and append its output
            for tool in function_calls:
                tool_name = tool.name
                tool_args = json.loads(tool.arguments)
                fun_to_ex = tool_map[tool_name]
                tool_output = fun_to_ex(**tool_args)

                input_messages.append({
                    "type": "function_call_output",
                    "call_id": tool.call_id,
                    "output": json.dumps(tool_output)
                })

        else:
            break

    return response.output_text
```
