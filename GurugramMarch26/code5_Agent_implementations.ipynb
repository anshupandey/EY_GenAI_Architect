## Agent Implementation with LangChain

```python
# initialize the llm
from langchain.chat_models import init_chat_model
model = init_chat_model("azure_openai:gpt-4o-mini")

from langchain_community.utilities import OpenWeatherMapAPIWrapper
weather = OpenWeatherMapAPIWrapper()
```

```python
# create a simple agent
from langchain.agents import create_agent
agent = create_agent(model =model,
                     tools=[weather.run],
                     system_prompt="You are a helpful assistant, choose th tool if you need appropriately")
agent.invoke({"messages":[{"role":"user","content":"what is AI?"}]})
```

```python
from langchain_tavily import TavilySearch
tavily_tool = TavilySearch(max_results=5, topic="general",)

from langchain_community.tools import WikipediaQueryRun
from langchain_community.utilities import WikipediaAPIWrapper
wikipedia = WikipediaQueryRun(api_wrapper=WikipediaAPIWrapper())
```

```python
# create a simple agent
from langchain.agents import create_agent
agent = create_agent(model =model,
                     tools=[weather.run,wikipedia,tavily_tool],
                     system_prompt="You are a helpful assistant, choose th tool if you need appropriately, review your responses, ALso do planning the beginining, show your plan, act upon it and sobserve, change if needed. ")
response = agent.invoke({"messages":[{"role":"user","content":"Perform a detailed research on future of software and indian software cities, their current weather and future prospects. "}]})
```

```python
for msg in response['messages']:
    if msg.type in ['ai','humna']:
        print("==="*20)
        print('\n')
        print(msg.content)
    else:
        print(msg)
```
